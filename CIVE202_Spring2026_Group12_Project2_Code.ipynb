```python
import pandas as pd
import numpy as np
```

```python
# define a function to calculate water weight
def calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E):
    
    # Our first local variable: total cementitious weight
    total_cementious = cement_A + fly_ash_B + silica_fume_C + other_SCM_D
    
    # Our second local variable: water weight
    water_weight_Q = total_cementious * wc_ratio_E

    # this returns water weight
    return water_weight_Q
uww = 62.4
```

```python
# Define a function to calculate the volume of cement (R)

def cementious_volumes_R(cement_A, sg_J):
    volume_R = cement_A / (sg_J * uww)
    return volume_R
```

```python
def fa_vol_S(fa_B, sg_K):
    volume_S = fa_B / (sg_K * uww)
    return volume_S
```

```python
def sf_vol_T(sf_C, sg_L):
    volume_T = sf_C / (sg_L * uww)
    return volume_T
```

```python
def SCM_vol_U(scm_D, sg_M):
    volume_U = scm_D / (sg_M * uww)
    return volume_U
```

```python
def AC_V(Air_F):
    vol_V = (Air_F / 100)*27
    return vol_V
```

```python
def water_vol_W(ww_Q):
    volume_W = ww_Q / 62.4
    return volume_W
```

```python
def total_aggregates_X(R, S, T, U, V, W):
    X = 27 - (R + S + T + U + V + W)
    return X
```

```python
def fine_agg_weight_Y(G, N, X):
    return 62.4*(G/100) * N * X
```

```python
# Function for coarse aggregate weight (Z)
def calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, volume_X):
    return uww * (percent_coarse_H / 100) * sg_coarse_O * volume_X
```

```python
# Function for other aggregate weight (AA)
def calculate_other_aggregate_AA(percent_other_I, sg_other_P, volume_X):
    return uww * (percent_other_I / 100) * sg_other_P * volume_X
```

```python
# General info
project_no = int(input("Enter project number: ")) 
concrete_class = input("Enter class of concrete: ")

# Cementitious material inputs
cement_A = float(input("Enter weight of cement: ")) 
fly_ash_B = float(input("Enter weight of fly ash: ")) 
silica_fume_C = float(input("Enter weight of silica fume: ")) 
other_SCM_D = float(input("Enter weight of other SCM: ")) 

# Design parameters
concrete_class = input("Enter class of concrete: ")
wc_ratio_E = float(input("Enter water cement ratio: ")) 
Air_F = float(input("Enter target air content: "))

# Aggregate proportions
fine_percent_G = float(input("Enter fine percent aggregate: ")) 
percent_coarse_H = float(input("Enter coarse percent aggregate: "))
percent_other_I = float(input("Enter other percent aggregate: "))


# Specific gravities
sg_cement_J = float(input("Enter specific gravity of cement: ")) 
sg_fly_ash_K = float(input("Enter specific gravity of fly ash: ")) 
sg_silica_fume_L = float(input("Enter specific gravity of silica fume: ")) 
sg_other_scm_M = float(input("Enter specific gravity of other SCM: ")) 
sg_fine_N = float(input("Enter specific gravity fine aggregates: ")) 
sg_coarse_O = float(input("Enter specific gravity coarse aggregates: ")) 
sg_other_P = float(input("Enter specific gravity other aggregates: ")) 


# A Print message to confirm all inputs have been collected
print("\nAll user inputs collected successfully.")
```

```python
# Step 1: Water weight (Q)
Q = calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_SCM_D, wc_ratio_E)
# Step 2: Cementitious volumes (R, S, T, U)
R = cementious_volumes_R(cement_A, sg_cement_J)
S = fa_vol_S(fly_ash_B, sg_fly_ash_K)
T = sf_vol_T(silica_fume_C, sg_silica_fume_L)
U = SCM_vol_U(other_SCM_D, sg_other_scm_M)
# Step 3: Air and water volumes (V, W)
V = AC_V(Air_F)
W = water_vol_W(Q)
# Step 4: Total aggregate volume (X)
X = total_aggregates_X(R, S, T, U, V, W)
# Step 5: Aggregate weights (Y, Z, AA)

Y = fine_agg_weight_Y(fine_percent_G, sg_fine_N, X)
Z = calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, X)
AA = calculate_other_aggregate_AA(percent_other_I, sg_other_P, X)
```
